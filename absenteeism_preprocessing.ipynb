```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
raw_csv_data=pd.read_csv("Absenteeism_data.csv")
raw_csv_data
```

```python
df=raw_csv_data.copy()
```

```python
df.info()
```

```python
pd.options.display.max_columns=None
pd.options.display.max_rows=20
#print(df)
```

```python
df=df.drop("ID",axis=1)
```

```python
#print(df)
```

```python
reason=pd.get_dummies(df['Reason for Absence'],drop_first=True)
```

```python
#print(reason)
```

```python
reason['check']=reason.sum(axis=1)
#print(reason)
```

```python

reason=reason.drop("check",axis=1)
reason1=reason.loc[:,1:14].max(axis=1)
reason2=reason.loc[:,15:17].max(axis=1)
reason3=reason.loc[:,18:21].max(axis=1)
reason4=reason.loc[:,22:  ].max(axis=1)


#reason
```

```python
df=df.drop("Reason for Absence",axis=1)
df=pd.concat([df,reason1,reason2,reason3,reason4],axis=1)
#print(df)
```

```python
df.columns.values
```

```python
renamed_columns=['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'reason1', 'reason2', 'reason3', 'reason4',]
df.columns=renamed_columns
reordered_columns=[ 'reason1', 'reason2', 'reason3', 'reason4','Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df=df[reordered_columns]
#df
```

CHECKPOINT

reasons modified and grouped

```python
df_reason_mod=df.copy()
#df_reason_mod
```

```python
print(type(df_reason_mod['Date'][1]))
df_reason_mod['Date']=pd.to_datetime(df_reason_mod['Date'],format="%d/%m/%Y")
type(df_reason_mod['Date'][1])
```

```python
df_reason_mod['Date'][5].weekday()+1
df_reason_mod['Date'][5].month
```

```python
month_list=[]
week_list=[]
for i in range(df_reason_mod.shape[0]):
    month_list.append(df_reason_mod['Date'][i].month)
    week_list.append(df_reason_mod['Date'][i].weekday()+1)
df_reason_mod['month']=month_list
df_reason_mod['week_day']=week_list
df_reason_mod.columns.values
```

```python
df_reason_mod=df_reason_mod[['reason1', 'reason2', 'reason3', 'reason4', 'Date','month','week_day',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']]
#df_reason_mod.head(10)
```

```python
df_reason_mod=df_reason_mod.drop('Date',axis=1)
#df_reason_mod
```

CHECKPOINT-DATE TRANSFORMED 
month and week replaces date

```python
df_date_transformed=df_reason_mod.copy()
#df_date_transformed.head()
```

```python
df_date_transformed['Education'].value_counts()
education_transformed=np.where(df_date_transformed['Education']==1,1,0)
df_date_transformed['Education']=education_transformed
```

```python
#df_date_transformed
```

```python
scaler=StandardScaler()
scale=scaler.fit(df_date_transformed.iloc(axis=1)[6:11])
scaled_values=scale.fit_transform(df_date_transformed.iloc(axis=1)[6:11])
df_date_transformed[df_date_transformed.iloc(axis=1)[6:11].columns.values]=scaled_values
```

Now our data is finally preprocessed

```python
df_preprocessed=df_date_transformed.copy()
```
